## Chart Rquirement

## **Total Sales by Fat content**

```python
Sales_by_fat=Data.groupby('Item Fat Content')['Sales'].sum()
plt.pie(Sales_by_fat,labels=Sales_by_fat.index,
        autopct='%.1f%%',
        startangle=90)
plt.title('Sales by Fat Content')
plt.show()
```

```python
sales_by_types=Data.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
bars=plt.bar(sales_by_type.index,sales_by_types.values)
plt.xticks(rotation=-90)
plt.xlabel('Item Types')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')
plt.tight_layout()
plt.show()
```

## **Fat content of outlet by total_sale**

```python
Fat_content_by_Outlet=Data.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()
Fat_content_by_Outlet=Fat_content_by_Outlet['Regular','Low Fat']
ax=Fat_content_by_Outlet.plot(kind='bar', figsize(8, 5), title='Outlet tier by Item Fat Content')
plt.xlabel('Outlet location tier')
plt.ylabel('Total Sales')
plt.title('item Fat content')
plt.tight_layout()
plt.show()
```
